```python
from scipy.optimize import minimize
```

```python
rend=rend.loc[:, rend.columns != 'MEXBOL INDEX']
```

```python
def port_vol(w,r):
    s_p=np.sqrt(w.T@(rend.cov()*252)@w)
    return s_p

N, M  = rend.shape
w0 = np.random.randn(M)

def sum_weights(weights):
    return weights.sum() -1

constraints = [
    {"type":"eq", "fun":sum_weights}
]
port0 = minimize(port_vol, w0, constraints=constraints, args=rend)
port0
```

```python
w=port0.x
w
```

```python
def port_rend(w, r_capm):
    E_p =(r_capm @ w)
    return E_p
port_rend(w, rend_CAPM)
```

```python
rmin = rend_CAPM.min()
rmax = rend_CAPM.max()
```

```python
rend_max = np.linspace(rmin,rmax)
```

```python
def rend_esperado(w, E):
    return port_rend(w, rend_CAPM) - E # == 0
```

```python
pesos_optimos = []

for e in rend_max:
    constraints = [
        {"type":"eq", "fun":sum_weights},
        {"type":"eq", "fun":lambda w: rend_esperado(w, e)}
    ]
    port1 = minimize(port_vol, w0, constraints=constraints, args=rend)
    w_opt = port1.x
    pesos_optimos.append(w_opt)
```

```python
r_opt = []
v_opt = []
for w in pesos_optimos:
    r_opt.append(port_rend(w,rend_CAPM))
    v_opt.append(port_vol(w,rend))
r_opt = np.array(r_opt)
v_opt = np.array(v_opt)

plt.scatter(v_opt,r_opt, c=r_opt/v_opt)
plt.colorbar();
```

```python
def min_func_sharpe(weights):
        return - (port_rend(weights, rend_CAPM) - tlr)/ port_vol(weights,rend)

constraints = [
    {"type":"eq", "fun":sum_weights}
]
f_sharpe = minimize(min_func_sharpe,w0, constraints=constraints)
```

```python
w_sharpe = f_sharpe.x
w_sharpe
```

```python
e_sharpe = port_rend(w_sharpe, rend_CAPM)
vol_sharpe = port_vol(w_sharpe, rend)
```

```python
sigma_c = np.linspace(0,0.25)
sharpe = tlr + sigma_c*(e_sharpe - tlr)/vol_sharpe
plt.plot(sigma_c, sharpe, linestyle="--",c="tab:orange", label="Capital Allocation Line")
plt.plot(v_opt, r_opt)
plt.legend(fontsize=12);
```

Backpropagation es un algoritmo que funciona mediante la determinación de la pérdida (o error) en la salida y luego propagándolo de nuevo hacia atrás en la red. De esta forma los pesos se van actualizando para minimizar el error resultante de cada neurona. Este algoritmo es lo que les permite a las redes neuronales aprender.
